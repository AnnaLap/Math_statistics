```python
from math import e
from math import factorial 
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

### **Задача 1**. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.


```python
# Используем формулу Бернулли, т.к. вероятность попадания высокая и число выстрелов не слишком велико
res = combinations(100, 85) * (0.8**85) * (0.2**(100-85))
print(f'Вероятность, что стрелок из 100 раз попадет ровно 85 =' , round(res, 5))
```

### **Задача 2**. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. 
#### a) Какова вероятность, что ни одна из них не перегорит в первый день? 
#### b) Какова вероятность, что перегорят ровно две?

```python
# Используем формулу распределения Пуассона, т.к. вероятность перегорания лампочки очень низкая и число новых лампочек велико
# а) ни одна не перегорит
lambdya = 0.0004*5000
pa = (lambdya**0 * e**(-lambdya))/factorial(0)
print(f'Вероятность, что ни одна лампочка не перегорит в первый день =' , round(pa, 5))
# b) перегорят две
pb = (lambdya**2 * e**(-lambdya))/factorial(2)
print(f'Вероятность, что перегорят две лампочки =' , round(pb, 5))
```

### **Задача 3**. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

```python
# Используем формулу Бернулли, т.к. вероятность выпадения орла высокая, число подбрасываний мало
res = combinations(144, 70) * (0.5**70) * (0.5**(144-70))
print(f'Вероятность, что орел выпадет ровно 70 раз =' , round(res, 5))
```

### **Задача 4**. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. 
#### a) Какова вероятность того, что все мячи белые? 
#### b) Какова вероятность того, что ровно два мяча белые? 
#### c) Какова вероятность того, что хотя бы один мяч белый?

```python
# а) все белые (случится такое, если мы выберем сначала какой-то из 7 белых из 10 возможных, потом один из 6 белых из оставшихся 9 
# и один из 9 белых из 11 и один из 8 из 10 оставшихся)
pa = (7/10)*(6/9)*(9/11)*(8/10)
print(f'Вероятность, что все белые =' , round(pa, 5))
# b) ровно 2 белые (варианты: ЧЧ и ББ или ЧБ и ЧБ или БЧ и БЧ или ББ и ЧЧ или БЧ и ЧБ или ЧБ и БЧ)
p1 = (3/10)*(2/9)*(9/11)*(8/10)
p2 = (3/10)*(7/9)*(2/11)*(9/10)
p3 = (7/10)*(3/9)*(9/11)*(2/10)
p4 = (7/10)*(6/9)*(2/11)*(1/10)
p5 = (7/10)*(3/9)*(2/11)*(9/10)
p6 = (3/10)*(7/9)*(9/11)*(2/10)
pb = p1+p2+p3+p4+p5+p6
print(f'Вероятность, что ровно 2 белые =' , round(pb, 5))
# с) хотя бы один белый (от противного: все черные)
pc = 1 - (3/10)*(2/9)*(2/11)*(1/10)
print(f'Вероятность, что хотя бы один белый =' , round(pc, 5))
```
