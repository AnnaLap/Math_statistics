### **Задача 1**. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

```python
# математическое ожидание - середина отрезка, дисперсия по формуле
a = 200
b = 800
m = (a+b)/2
d = (b-a)**2/12
print(f'M(A) =', m)
print(f'D(A) =', d)
```

### **Задача 4**.  Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
### 1. больше 182 см?
### 2. больше 190 см?
### 3. от 166 см до 190 см?
### 4. от 166 см до 182 см?
### 5. от 158 см до 190 см?
### 6. не выше 150 см или не ниже 190 см?
### 7. не выше 150 см или не ниже 198 см?
### 8. ниже 166 см?
### Задачу можно решить двумя способами: без использования сторонних библиотек (numpy, scipy, pandas и пр.), а затем проверить себя с помощью встроенных функций
 

```python
from statistics import NormalDist
#1. так как 182см равно односигмовому интервалу (174+8), то Z = 1 и значение вероятности можем посмотреть по Z-таблице. 
# Так как нам нужны значения БОЛЬШЕ 182см, то нужно вычесть полученное по таблице значение из 1
P1 = 1 - .84134
print('1) Вручную: Р(Х>=182) =', round(P1, 5))
P1_auto = 1 - NormalDist().cdf((182 - 174)/8)
print('Програмно: Р(Х>=182) =', round(P1_auto, 5))
print('')

#2. Аналогично пункту 1, только для двухсигмового интервала
P2 = 1 - 0.97725
print('2) Вручную: Р(Х>=190) =', round(P2, 5))
P2_auto = 1 - NormalDist().cdf((190 - 174)/8)
print('Програмно: Р(Х>=190) =', round(P2_auto, 5))
print('')

#3. 166-190: 166 - минус односигмовый интервал (z = -1). Для 190 знаем что z = 2
# Так как нам нужны значения между 166см и 190см, то нужно вычесть из вероятности для 190 вероятность для 166
P3 = 0.97725 - 0.15866
print('3) Вручную: Р(166<Х<190) =', round(P3, 5))
P3_auto = NormalDist().cdf((190 - 174)/8) - NormalDist().cdf((166 - 174)/8)
print('Програмно: Р(166<Х<190) =', round(P3_auto, 5))
print('')

#4. 166-182: аналогично пункту 3, только между -1 и +1 сигмой
P4 = 0.84134 - 0.15866
print('4) Вручную: Р(166<Х<182) =', round(P4, 5))
P4_auto = NormalDist().cdf((182 - 174)/8) - NormalDist().cdf((166 - 174)/8)
print('Програмно: Р(166<Х<182) =', round(P4_auto, 5))
print('')

#5. 158-190: аналогично пункту 3, только между -2 и +2 сигмой
P5 = 0.97725 - 0.02275
print('5) Вручную: Р(158<Х<190) =', round(P5, 5))
P5_auto = NormalDist().cdf((190 - 174)/8) - NormalDist().cdf((158 - 174)/8)
print('Програмно: Р(158<Х<190) =', round(P5_auto, 5))
print('')

#6. Х<150 или Х>190: 150 - минус трехсигмовый интервал (z = -3). Для 190 знаем что z = 2
# Так как нам нужны значения или меньше 150 или больше 190 мы можем вычесть из 1 значения между этих точек
P6 = 1 - (0.97725 - 0.00135)
print('6) Вручную: Р(190<Х<150) =', round(P6, 5))
P6_auto = 1 - (NormalDist().cdf((190 - 174)/8) - NormalDist().cdf((150 - 174)/8))
print('Програмно: Р(190<Х<150) =', round(P6_auto, 5))
print('')

#7. Х<150 или Х>198: аналогично пункту 6 только с z1 = 3, z2 = -3
P7 = 1 - (0.99865 - 0.00135)
print('7) Вручную: Р(198<Х<150) =', round(P7, 5))
P7_auto = 1 - (NormalDist().cdf((198 - 174)/8) - NormalDist().cdf((150 - 174)/8))
print('Програмно: Р(198<Х<150) =', round(P7_auto, 5))
print('')

#8. X<166: z = -1
P8 = .15866
print('8) Вручную: Р(Х<166) =', round(P8, 5))
P8_auto = NormalDist().cdf((166 - 174)/8)
print('Програмно: Р(Х<166) =', round(P8_auto, 5))

```

### **Задача 5**. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
# Чтобы перевести в сигма размерность, воспользуемся известной формулой
x = 190
m = 178
d = 25
sko = d**(1/2)
z = (x - m)/sko
print(f'Количество сигм для 190см при мат. ожидании 178см =', z)
```
