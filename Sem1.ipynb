# Теория вероятностей и математическая статистика
## Cеминар 1

```python
from math import factorial 
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

### **Задача 1**.  Из колоды в 52 карты извлекаются случайным образом 4 карты. 
### a) Найти вероятность того, что все 4 карты – крести. 

```python
# вероятность будет равна отношению числа сочетаний выбора 4 карт крести из всех карт с мастью крести (4 из 13)
#  к числу сочетаний выбора 4 карт из общей колоды (4 из 52)
s4_13 = combinations(13, 4)
print(f'Число сочетаний выбора 4 карт из всех "крести" = ', s4_13)
s4_52 = combinations(52, 4)
print(f'Число сочетаний выбора 4 карт из всей колоды = ', s4_52)
res = s4_13/s4_52
print(f'Вероятность, что 4 карты - крести = ', round(res, 5))
```

### б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз. 

```python
# решим от противного, зная, что вероятность вытянуть хотя бды один туз = 1 - вероятность не вытянуть ни одного
# вероятность не вытянуть ни одного будет равна отношению числа сочетаний выбора 4 карт крести из колоды без тузов (4 из 48)
# к числу сочетаний выбора 4 карт из общей колоды (4 из 52)
s4_48 = combinations(48, 4)
print(f'Число сочетаний выбора 4 карт из всех кроме тузов = ', s4_48)
s4_52 = combinations(52, 4)
print(f'Число сочетаний выбора 4 карт из всей колоды = ', s4_52)
res = 1 - s4_48/s4_52
print(f'Вероятность, что среди 4-х карт окажется хотя бы один туз = ', round(res, 5))
```

### **Задача 2**. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно . Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
# Вероятность будет равна отношению вариантов верных кодов (он всего 1) к числу сочетаний три по десять (одновременное нажатие)
res = 1/combinations(10, 3)
print(f'Вероятность, ччто человек, не знающий код, откроет дверь с первой попытки = ', round(res, 5))
```

### **Задача 3**. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все 3 извлеченные детали окрашены?

```python
# вероятность будет равна отношению числа сочетаний выбора 3 деталей из окрашенных (3 из 9)
#  к числу сочетаний выбора 3 деталей из всех (3 из 15)
s3_9 = combinations(9, 3)
print(f'Число сочетаний выбора 3 деталей из окрашенных = ', s3_9)
s3_15 = combinations(15, 3)
print(f'Число сочетаний выбора 3 деталей из всех = ', s3_15)
res = s3_9/s3_15
print(f'Вероятность, что все 3 извлеченные детали окрашены = ', round(res, 5))
```

### **Задача 4**. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
# вероятность равна произведению вероятностей, что первый вытянутый выигрышный (2/100) и второй вытянутый из оставшихся выигрышный (1/99)
res = (2/100) * (1/99)
print(f'Вероятность, что оба билета выигрышные = ', round(res, 5))
```
